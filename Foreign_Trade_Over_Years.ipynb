```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('economy_foreign_trade.csv')
```

```python
# Pivot the data to get states as rows and years as columns
pivoted_df = df.pivot_table(index='State', columns='Year', values='Trade Value', aggfunc='sum')

# Create a custom colormap from light blue to dark blue
from matplotlib.colors import LinearSegmentedColormap

# Define a light to dark blue gradient
colors = ['#add8e6', '#87ceeb', '#4682b4', '#4169e1', '#0000cd', '#00008b', '#000080']  # Light to dark blue
n_bins = len(pivoted_df.columns)  # Number of years
cmap = LinearSegmentedColormap.from_list("Gradient_Blue", colors, N=n_bins)

# Plotting the horizontal stacked bar chart
plt.figure(figsize=(120, 80))
pivoted_df.plot(kind='barh', stacked=True, colormap=cmap)

# Setting the formatter directly using a lambda function
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'{x * 1e-12:.0f}T'))

plt.title('Trade Value per State by Year')
plt.ylabel('State')
plt.xlabel('Trade Value')
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.show()
```
