## What are the top 5 countries that Mexico buys from in the last decade?

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Trade Origins data files (10 years: 2014 - 2023) paths
trade_origins_2014_path = "datasources/International Purchases/Trade-Origins-2014.csv"
trade_origins_2015_path = "datasources/International Purchases/Trade-Origins-2015.csv"
trade_origins_2016_path = "datasources/International Purchases/Trade-Origins-2016.csv"
trade_origins_2017_path = "datasources/International Purchases/Trade-Origins-2017.csv"
trade_origins_2018_path = "datasources/International Purchases/Trade-Origins-2018.csv"
trade_origins_2019_path = "datasources/International Purchases/Trade-Origins-2019.csv"
trade_origins_2020_path = "datasources/International Purchases/Trade-Origins-2020.csv"
trade_origins_2021_path = "datasources/International Purchases/Trade-Origins-2021.csv"
trade_origins_2022_path = "datasources/International Purchases/Trade-Origins-2022.csv"
trade_origins_2023_path = "datasources/International Purchases/Trade-Origins-2023.csv"

# CSVs to pandas df
trade_origins_2014 = pd.read_csv(trade_origins_2014_path)
trade_origins_2015 = pd.read_csv(trade_origins_2015_path)
trade_origins_2016 = pd.read_csv(trade_origins_2016_path)
trade_origins_2017 = pd.read_csv(trade_origins_2017_path)
trade_origins_2018 = pd.read_csv(trade_origins_2018_path)
trade_origins_2019 = pd.read_csv(trade_origins_2019_path)
trade_origins_2020 = pd.read_csv(trade_origins_2020_path)
trade_origins_2021 = pd.read_csv(trade_origins_2021_path)
trade_origins_2022 = pd.read_csv(trade_origins_2022_path)
trade_origins_2023 = pd.read_csv(trade_origins_2023_path)

trade_origins_2014
```

```python
# Deleting unnecesarry columns
trade_origins_2014 = trade_origins_2014.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2015 = trade_origins_2015.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2016 = trade_origins_2016.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2017 = trade_origins_2017.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2018 = trade_origins_2018.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2019 = trade_origins_2019.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2020 = trade_origins_2020.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2021 = trade_origins_2021.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2022 = trade_origins_2022.drop(columns=["Continent ID","Continent","Country ID","Share"])
trade_origins_2023 = trade_origins_2023.drop(columns=["Continent ID","Continent","Country ID","Share"])

trade_origins_2014
```

```python
# Sorting by "Trade Value" and getting only the first 5 values (most important countries)
trade_origins_2014 = trade_origins_2014.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2015 = trade_origins_2015.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2016 = trade_origins_2016.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2017 = trade_origins_2017.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2018 = trade_origins_2018.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2019 = trade_origins_2019.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2020 = trade_origins_2020.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2021 = trade_origins_2021.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2022 = trade_origins_2022.sort_values(by=["Trade Value"],ascending=False).head()
trade_origins_2023 = trade_origins_2023.sort_values(by=["Trade Value"],ascending=False).head()

# Since the 5 top countries never change, we order alfabetically and then merge all in one dataframe
trade_origins_2014 = trade_origins_2014.sort_values(by=["Country"])
trade_origins_2015 = trade_origins_2015.sort_values(by=["Country"])
trade_origins_2016 = trade_origins_2016.sort_values(by=["Country"])
trade_origins_2017 = trade_origins_2017.sort_values(by=["Country"])
trade_origins_2018 = trade_origins_2018.sort_values(by=["Country"])
trade_origins_2019 = trade_origins_2019.sort_values(by=["Country"])
trade_origins_2020 = trade_origins_2020.sort_values(by=["Country"])
trade_origins_2021 = trade_origins_2021.sort_values(by=["Country"])
trade_origins_2022 = trade_origins_2022.sort_values(by=["Country"])
trade_origins_2023 = trade_origins_2023.sort_values(by=["Country"])

trade_origins_2014
```

```python
# Combine all years trade values into a signle dataframe
trade_origins = pd.concat([i.set_index("Country") for i in [trade_origins_2014, trade_origins_2015, 
                                                     trade_origins_2016, trade_origins_2017,
                                                     trade_origins_2018, trade_origins_2019,
                                                     trade_origins_2020, trade_origins_2021,
                                                     trade_origins_2022, trade_origins_2023]],axis=1)

trade_origins
```

```python
# Renaming the columns
trade_origins.columns = ["2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023"]
x_axis = np.arange(1,11,1)

# Transpose for plotting
trade_origins = trade_origins.T
trade_origins
```

```python
# Plotting a stackplot
labels = ["Germany","South Korea","Japan","China","United States"]
plt.stackplot(x_axis, trade_origins["Germany"], trade_origins["South Korea"], trade_origins["Japan"], 
              trade_origins["China"], trade_origins["United States"], alpha=0.9, colors =["yellow","green","gray","red","blue"])

years = ["2014", "2015", "2016", "2017", "2018", "2019", "2020", "2021", "2022", "2023"]
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, years)

plt.legend(labels=labels,loc="upper left")
plt.ylabel("USD (Billions)")
plt.title("Mexico's International purchases over the last decade")

# Plotting a vertical line in 2020 to simbolize covid
plt.axvline(x = 7, color="black")

plt.grid()
plt.show()
```

#### Major findings
The 5 countries we purchase the most from did not change over time, with the United States and China on<br>
top with a significant difference.

In 2018, Mexico held elections for president and this could explained a rise on purchases, beginning from<br>
2017. 

Then in 2020, due to COVID-19 it was expected that China ranked lower, but that did not happened. Although <br>
there was less spending in general. 

Lastly in 2021-2022 the metrics spiked as globally we recovered from the pandemic and we experienced <br>
political stability (presidency on midterm). 


```python
# Plotting pie chart of sum all years
china_sum = trade_origins["China"].sum()
germany_sum = trade_origins["Germany"].sum()
japan_sum = trade_origins["Japan"].sum()
south_korea_sum = trade_origins["South Korea"].sum()
united_states_sum = trade_origins["United States"].sum()

pies = ["China","Germany","Japan","South Korea","United States"]
sums = [china_sum,germany_sum,japan_sum,south_korea_sum,united_states_sum]
colors = ["red","yellow","gray","green","blue"]
explode = (0,0,0,0,0.02)

# Tell matplotlib to create a pie chart based upon the above data
plt.pie(sums, explode=explode, labels=pies, colors=colors,
        autopct="%1.1f%%")
plt.title("Mexico's Gross Int'l Purchases since 2013")
plt.show()
```

#### Major findings
After adding up all the sums of the past decade, it was not only clear that the United States lead <br>
the list, but it's actually bigger than the sum that all the other countries, which unfortunately that <br>
means that Mexico is strongly dependent on USA's economy.


In this graph we observe the sum of all the purchases in the last decade. Mexico depends heavily on <br>
the United States economy, as they are our biggest provider of oils and petroleum. This could be considered <br>
negative as we should diversify our spending. 

China has been playing a greater role in the country's economy as they introduced massively new OEMs <br>
(original equipment manufacturers, car brands) so vehicle parts and accessories, as well as electronic <br>
integrated circuits, have been our top product bought. 

We may see a bigger participation from China and then we could shorten the gap between our first two places, <br>
purchasing-wise.
